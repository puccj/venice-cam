# Plot the RMSE of the counting

```python
import csv
import matplotlib.pyplot as plt

# File path
file_path_train = '/home/daniele/venice-cam/output/all_crowd_count_distance_train.csv'
file_path_val = '/home/daniele/venice-cam/output/all_crowd_count_distance_val.csv'

# Lists to store data
min_distance_train = []
mean_square_error_train = []

min_distance_val = []
mean_square_error_val = []

# Read the CSV file
with open(file_path_train, 'r') as csvfile:
  reader = csv.DictReader(csvfile)
  for row in reader:
    min_distance_train.append(float(row['Min distance']))
    mean_square_error_train.append(float(row['Mean square error'])**0.5)

with open(file_path_val, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        min_distance_val.append(float(row['Min distance']))
        mean_square_error_val.append(float(row['Mean square error'])**0.5)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(min_distance_train, mean_square_error_train, marker='o', linestyle='-', color='b')
plt.plot(min_distance_val, mean_square_error_val, marker='o', linestyle='-', color='r')
plt.title('Root Mean Square Error over Min Distance', fontsize=16)
plt.xlabel('Min Distance (px)', fontsize=14)
plt.ylabel('Root Mean Square Error', fontsize=14)
plt.legend(['Train', 'Validation'], fontsize=12)
plt.grid(True)
plt.show()
```

# Plot the data counting with min_distance=0.2

```python
import csv
import matplotlib.pyplot as plt

# File path
file_path_train = 'output/crowd_count/0.20-train.csv'
file_path_val = 'output/crowd_count/0.20-val.csv'

# Lists to store data
count_train = []
true_train = []

count_val = []
true_val = []

# Read the CSV file
with open(file_path_train, 'r') as csvfile:
  reader = csv.DictReader(csvfile)
  for row in reader:
    count_train.append(float(row['Count']))
    true_train.append(float(row['True']))

with open(file_path_val, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        count_val.append(float(row['Count']))
        true_val.append(float(row['True']))

# sort data according to true count
true_train, count_train = zip(*sorted(zip(true_train, count_train)))
true_val, count_val = zip(*sorted(zip(true_val, count_val)))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(true_train, count_train, marker='o', linestyle='', color='b')
plt.plot(true_val, count_val, marker='o', linestyle='', color='r')
plt.plot([min(true_train + true_val), max(true_train + true_val)], [min(true_train + true_val), max(true_train + true_val)], linestyle='--', color='gray')
plt.title('Predicted vs True Count', fontsize=16)
plt.xlabel('True Count', fontsize=14)
plt.ylabel('Predicted Count', fontsize=14)
plt.legend(['Train', 'Validation'], fontsize=12)
plt.grid(True)
plt.show()
```
